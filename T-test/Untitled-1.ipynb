```python
import random
import numpy
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def get_sample_statistics_from_small_sample(sample):
    sample_statistics={}
    mean=sum(sample)/len(sample)
    mean_with_np=numpy.mean(sample)
    SD=numpy.std(sample)
    
    N=len(sample)
    correction_factor=(N/(N-1))**0.5
    SD_plus=SD*correction_factor

    sample_statistics['mean']=mean
    sample_statistics['mean_with_np']=mean_with_np
    sample_statistics['SD']=SD
    sample_statistics['SD_plus']=SD_plus
    sample_statistics['SE_for_z']=SD*(N**0.5)/N
    sample_statistics['SE_for_t']=SD_plus*(N**0.5)/N
    return sample_statistics
```

```python
sample=[78,83,68,72,88]
result=get_sample_statistics_from_small_sample(sample)
expected=70
observed=result['mean']
SE_for_t_test=result['SE_for_t']
dof=len(sample)-1
t_test=(observed-expected)/SE_for_t_test
t_test,dof
```
